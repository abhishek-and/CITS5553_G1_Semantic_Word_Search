```python
from keybert import KeyBERT

# Input text
doc = " CCTV installation cctv  having Goods and Services from 2023-03-10 to 2023-04-20 cost from $5000 to $7000"

# Initialize KeyBERT model
kw_model = KeyBERT()

# Extract keywords
keywords = kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 2), stop_words=None)

# Filter keywords based on length
filtered_keywords = [keyword for keyword, _ in keywords if len(keyword.split()) > 1]

print("Query - ",doc)
print("Extracted words -")
# Print filtered keywords
for kw in filtered_keywords:
    print(kw)
```
